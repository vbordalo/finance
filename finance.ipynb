```python
df_DJI = web.DataReader('^DJI', 'yahoo', start, end) # Dow Jones Industrial Average (^DJI)
```

```python
df_DJI['MACD'] = df_DJI['Adj Close'].rolling(window=100).mean()
```

```python
plt.figure(figsize = (16, 12))

ax1 = plt.subplot2grid((6,1), (0,0), rowspan=3, colspan=1)
ax2 = plt.subplot2grid((6,1), (3,0), rowspan=1, colspan=1, sharex=ax1)

ax1.plot(df_DJI['Adj Close'])
ax1.plot(df_DJI['MACD'])
ax1.legend()
ax2.bar(df_DJI.index, df_DJI['Volume'])

plt.show()
```
